```python
import pandas as pd
pd.set_option('display.max_columns',10,'display.width',1000)
from sklearn.datasets import load_boston
boston_dataset = load_boston()
print(boston_dataset)
```

```python
#X is data (input)
#Y is target(output)
df_X = pd.DataFrame(boston_dataset.data)
df_Y = pd.DataFrame(boston_dataset.target)
columns = boston_dataset.feature_names #columns containg all names of columns
print(df_X)
print("%%"*45)
print(df_Y)
print("%%"*45)
print(columns)
```

```python
#Converting into dataframe
boston_df = pd.DataFrame(boston_dataset.data)
print(boston_df.head())
boston_df.columns = boston_dataset.feature_names
print(boston_df.head())
print("Shape of Dataset:",boston_df.shape)
```

```python
boston_df.describe()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Traing and Testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df_X,df_Y,test_size=0.33,random_state= 42)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(Y_test)
```

```python
print("Slope:",model.coef_)
print("Y-intercept:",model.intercept_)
print("Model Score:",model.score(X_test,Y_test))
```
